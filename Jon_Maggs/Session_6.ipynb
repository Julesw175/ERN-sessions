```python
#Session 5 recap
#Read in house prices from CSV
#Convert perdio to dattime
#Make a list with the highest, lowest, mean and median values
#Write a loop to print each value of that list seperately

import pandas as pd

df = pd.read_csv('/workspaces/ERN-sessions/data/1980 2023 average house prices.csv')

df.head()
```

```python
df['Period'] = pd.to_datetime(df['Period'],format= "%Y-%m")

max_price, min_price, mean_price, median_price = (
    df['Average price All property types'].max(),
    df['Average price All property types'].min(),
    df['Average price All property types'].mean(),
    df['Average price All property types'].median(),
)

price_summary = [max_price, min_price, mean_price, median_price]

for info in price_summary:
    print(info)

```

```python
#Start of session
# Equals ==
# Not Equals !=
# less then <
# greater than >
# greater or equal >=
# less or eqaul <=
# not ~

print((1 > 2) & (2 == 2))
print((1 > 2) | (2 == 2))
print((3 > 2) | (2 == 2))
print((3 > 2) & (2 == 2))
```

```python
print((10 > 12) & (12 > 10)) #fails an and condition
print((10 > 12) | (12 > 10)) #passes an or condition
```

```python
df_under_mean_median = df[(df['Average price All property types'] < df['Average price All property types'].mean()) &
                          (df['Average price All property types'] < df['Average price All property types'].median())]
df_under_mean_median
```

```python
over_mean = df['Average price All property types'] > df['Average price All property types'].mean()
under_median = df['Average price All property types'] < df['Average price All property types'].median()

condition = over_mean | under_median
df_over_under = df[~condition]
```

```python
percent_less_mean = df['Percentage change (monthly) All property types'] < df['Percentage change (monthly) All property types'].mean()
percent_less_median = df['Percentage change (yearly) All property types'] < df['Percentage change (yearly) All property types'].median()

condition1 = percent_less_mean & percent_less_median
over_under = df[condition1]
over_under
```
