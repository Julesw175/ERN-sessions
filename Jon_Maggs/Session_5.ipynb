```python
#RECAP - Data Frames

import pandas as pd

child_dict = {
    'Child_ID': ['child1', 'child2', 'child3', 'child4'],
    'Gender': [0, 1, 2, 7],
    'DOB': ['05/05/2015', '12/12/2020', '01/01/2000', '09/12/2017'],
    'Ethnicity': ['BANG', 'WOTH', 'MOTH', 'WIRI'],
    'PLACE': ['B1','T1','X2','W1']
}

child_data = pd.DataFrame(child_dict)

child_data['Gender'] = child_data['Gender'].astype(int)
child_data['DOB'] = pd.to_datetime(child_data['DOB'],format= '%d/%m/%Y')

child_data.info()




```

```python
mask = child_data['Gender'].isin([0, 1, 2])
wrong_gender_codes = child_data[~mask]
print(wrong_gender_codes)

incorrect_eth = child_data[~child_data['Ethnicity'].isin(['WIRI', 'MOTH', 'WOTH'])]
print(incorrect_eth)

correct_eth = child_data[child_data['Ethnicity'] != 'BANG']
print(correct_eth)

```

```python
corporation = [{'Name':'Will LP',
                'Salary':'10000',
                'Job Title':'Employee'},
                {'Name':'Annie LP',
                'Salary':'15000',
                'Job Title':'CEO'},
                {'Name':'Naiomi LP',
                'Salary':'90000',
                'Job Title':'VP'},
                {'Name':'Luke V',
                'Salary':'45000',
                'Job Title':'Teach Lead'},
                {'Name':'Matt K',
                'Salary':'30000',
                'Job Title':'Comms'},
                {'Name':'Hannah K',
                'Salary':'38000',
                'Job Title':'HR'},]

corp_df = pd.DataFrame(corporation)


corp_df['Date Started'] = ['05/12/1993', '01/01/2000', '01/01/2020', '23/12/1989', '10/10/2010', '17/12/2021']
corp_df['Salary'] = corp_df['Salary'].astype(int)

corp_df['Date Started'] = pd.to_datetime(corp_df['Date Started'], format='%d/%m/%Y')

corp_df
```

```python
# Equals ==
# Not Equals !=
# less then <
# greater than >
# greater or equal >=
# less of eqaul <=
# not ~

mask = corp_df['Salary'] > 40000

over_40k = corp_df[mask]
under_40k = corp_df[~mask]

under_40k
```

```python
mask = corp_df['Salary'] > corp_df['Salary'].mean()
over_mean = corp_df[mask]
over_mean
```

```python
spreads = ['butter', 'jam', 'peanut butter']
toppings = ['ham', 'cheese', 'sardines']
filling = []
combinations = []

mixins = []
sandwhiches = {}

for spread in spreads:
    for topping in toppings:
        mixins.append(f'{spread} {topping}')
    
    sandwhiches[spread] = mixins
    mixins = []

combinations_df = pd.DataFrame(sandwhiches)
combinations_df['topping'] = toppings
combinations_df.set_index('topping', inplace=True)

combinations_df
```
