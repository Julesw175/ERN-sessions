```python
springfield_jobs_dict = [{'Name':'Homer Simpson',
                  'Job Title':'Nuclear Safety Engineer',
                  'Salary($)':"25000"},
                  {'Name':'Joe Quimby',
                  'Job Title':'Mayor',
                  'Salary($)':"75000"},
                  {'Name':'Kent Brockman',
                  'Job Title':'News Anchor',
                  'Salary($)':"150000"},
                  {'Name':'Fat Tony',
                  'Job Title':'Mob Boss',
                  'Salary($)':"1000000"},
                  {'Name':'Apu Nahasapeemapetilon',
                  'Job Title':'Kwick-E_Mart Manager',
                  'Salary($)':"18000"},]

springfield_jobs_df = pd.DataFrame(springfield_jobs_dict)
springfield_jobs_df

```

```python
#print(springfield_jobs_df.info())
print(springfield_jobs_df.dtypes)
#print(springfield_jobs_df.head(2))
#print(springfield_jobs_df.tail(1))
```

```python
springfield_jobs_df['Salary($)'] = springfield_jobs_df['Salary($)'].astype(int)
springfield_jobs_df.dtypes
```

```python
salary_total = springfield_jobs_df['Salary($)'].sum()
salary_mean = springfield_jobs_df['Salary($)'].mean()
sal_min_max = f"salary minimum: {springfield_jobs_df['Salary($)'].min()}, salary maximum: {springfield_jobs_df['Salary($)'].max()}"
print(sal_min_max)

# Others .cumsum, .min, .max
```

```python
springfield_jobs_df['Date Started'] = ['05/12/1993', '01/01/2000', '01/01/2020', '01/01/1991', '04/03/1996']
springfield_jobs_df
```

```python
inflation = 1.067

springfield_jobs_df['Salary($)'] = springfield_jobs_df['Salary($)'] * inflation

springfield_jobs_df['Pay After Bonus'] = springfield_jobs_df['Salary($)'] + 500
springfield_jobs_df
```

```python
springfield_jobs_df['Date Started'] = pd.to_datetime(springfield_jobs_df['Date Started'], format='%d/%m/%Y')
springfield_jobs_df
```

```python
springfield_jobs_df['DOB'] = ['12/05/1965', '01/02/1970', '29/04/1990' , '03/09/1952', '27/08/1973']
springfield_jobs_df['DOB'] = pd.to_datetime(springfield_jobs_df['DOB'], format='%d/%m/%Y')
springfield_jobs_df['Time Served'] = pd.to_datetime('06/11/2023', format='%d/%m/%Y') - springfield_jobs_df['Date Started']
springfield_jobs_df['Time Served'] = springfield_jobs_df['Time Served'] / pd.Timedelta('365 days')
springfield_jobs_df['Time Served'] = springfield_jobs_df['Time Served'].round().astype('int', errors='ignore')
springfield_jobs_df

```
