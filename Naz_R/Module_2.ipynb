```python
child_identifiers = pd.DataFrame({'child_id':['child1', 'child2', 'child3', 'child4'],
              'age_first_contact':[6, 10, 12, 13], 
              'gender':['M', 'M', 'M', 'F']})

print (child_identifiers)

nhs_nums = pd.DataFrame({'child_id':['child1', 'child2', 'child3', 'child4'],
              'nhs_numbers':['nhs1', 'nhs2', 'nhs3', 'nhs4']})
print (nhs_nums)

ident_nhs = child_identifiers.merge(nhs_nums, how='left', left_on='child_id', right_on='child_id')

print(ident_nhs)
```

```python
cp_plans = pd.DataFrame({'child_id':['child1', 'child4'],
                        'cp_start_date':['01/01/2023', '01/01/2000']})

df_left = ident_nhs.merge(cp_plans, how='left', on='child_id')
print (df_left)
```

```python
df_under_mean = df[df['Average price All property types'] < df['Average price All property types'].median()]
df_under_mean
```

```python
df_under_mean_median = df[(df['Average price All property types'] < df['Average price All property types'].mean()) & (df['Average price All property types'] < df['Average price All property types'].median())]

df_under_mean_median
```

```python
over_mean = df['Average price All property types'] > df['Average price All property types'].mean()
under_median = df['Average price All property types'] < df['Average price All property types'].median()
print (over_mean)
print(under_median)

condition = over_mean & under_median
print(condition)
```

```python
df_over_under = df[condition]
df_over_under
```

```python
condition_two = over_mean | under_median
condition_two
```

```python
under_mean = df['Percentage change (monthly) All property types'] < df['Percentage change (monthly) All property types'].mean()
also_under_median = df['Percentage change (yearly) All property types'] < df['Percentage change (yearly) All property types'].median()

final_condition = under_mean & also_under_median
df_change = df[final_condition]
df_change
```
