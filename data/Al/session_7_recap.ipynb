```python
# Initialize the sum to zero
sum_of_odds = 0
up_to_100 = list(range(0,101))

# Iterate through numbers from 0 to 100
for number in up_to_100:
    # Check if the number is odd
    if number % 2 != 0:
        # Add the odd number to the sum
        sum_of_odds += number

# Print the result
print("Sum of odd numbers between 0 and 100:", sum_of_odds)
```
