```python
#Exercise 1 - create a list of 4 employees with name (space between forename and surname),
#salary and job title. Then convert that to a dataframe and print it. 
#Step 1 - create list of employees
list_of_emps = [{"Name":"Danny DeVito",
                  "Salary":50000.00,
                  "Job Title":"Twin"},
                  {"Name":"Arnold Schwarzanegger",
                  "Salary":100000.00,
                  "Job Title":"Mr Olympia"},
                  {"Name":"Sylvester Stallone",
                  "Salary":150000.00,
                  "Job Title":"Contender"},
                  {"Name":"Jackie Chan",
                  "Salary":200000.00,
                  "Job Title":"Stuntman"}]
#Step 2 - append to a dataframe
df_emps = pd.DataFrame(list_of_emps)
#Step 3 - print the dataframe
df_emps
```

```python
print(df_emps.info())
print(df_emps['Salary'].sum())
print(df_emps['Salary'].mean())
print(df_emps['Salary'].cumsum())
```
