```python
import itertools
import pandas as pd

# Given lists
spreads = ['butter', 'jam', 'peanut butter']
toppings = ['ham', 'cheese', 'sardines','coleslaw']

# Generate all possible combinations of the lists spreads and toppings
combinations = list(itertools.product(spreads, toppings))

# Create a DataFrame with the desired format
df = pd.DataFrame(index=toppings, columns=spreads)

# Fill in the DataFrame with the combinations
for spread, topping in combinations:
    df.at[topping, spread] = f"{spread} and {topping}"

# Print the DataFrame
df
```
